# 2. plot data

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
##rolling移动窗口，移动取值，往前取，如[1,2,3,4,5]，rolling(3),对于1和2，前面不足三个数字，则取值为NaN，对于3而言，取值为[1,2,3]
df['20d'] = np.round(df['Open'].rolling(20).mean(),2) ##calculate 20days average
df['40d'] = np.round(df['Open'].rolling(40).mean(),2) ##calculate 40days average
plt.figure(figsize=(16,5))
plt.grid(True)
plt.plot(df[['Open','20d','40d']])
plt.show()
```

# 5. Machine Learning for stocks price forecasting

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
start = datetime.strptime('2022-01-1','%Y-%m-%d')
end = datetime.strptime('2023-07-25','%Y-%m-%d')
stock = yf.download(tickers='0066.HK',start=start,end=end)
stock
```

根据今天的价格，预测明天的价格，那么我们需要准备的数据为：X=今日价格，Y=明日价格

```python
today = df['Open'].iloc[:len(df)-1].reset_index(drop=True)
tomorrow = df['Open'].iloc[1:].reset_index(drop=True)
data = pd.concat([today,tomorrow],axis=1,ignore_index=True)
data.columns = ['today','tomorrow']
data
```

```python
! pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error,r2_score
```

准备训练集和测试集

```python
today = np.array(today).reshape(-1,1)
tomorrow = np.array(tomorrow)
X_train,X_test,y_train,y_test = train_test_split(today,tomorrow,test_size=0.3)
```

训练模型并做预测

```python
model_lr = linear_model.LinearRegression()
model_lr.fit(X_train,y_train)
y_pred = model_lr.predict(X_test)
```

画出结果

```python
res = pd.DataFrame()
res['X_test'] = X_test.reshape(1,-1)[0]
res['y_test'] = y_test
res['y_pred'] = y_pred
res
```

```python
plt.figure(figsize=(16,6))
plt.plot(res[['y_test','y_pred']])
```

计算误差

```python
res_msr = mean_squared_error(y_test,y_pred)
res_r2_score = r2_score(y_test,y_pred)
print(f'mean square error= {res_msr} and r2_score = {res_r2_score}')
```
