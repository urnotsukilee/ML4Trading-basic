1. 使用yfinance获取股票价格（主要是美股和港股，深沪目前不在此API中）

```python
import yfinance as yf
```

```python
tickers = ['2252.HK','IBM']
df = yf.download(tickers,period='60d',interval='15m')
df
```

```python
df.columns
```

```python
df.loc[:,[('Close','IBM')]].dropna()
```
